```python
car_sales_missing_path = '/home/rujool/ML/datasets/car-sales-missing-data.csv'
csm = pd.read_csv(car_sales_missing_path)
csm.head()
```

```python
csm.Colour = csm.Colour.str.lower() # changing Colour column to lower case
csm.head()
```

```python
#suppose we fill Odometer column missing values with average of the rest
csm['Odometer'].fillna(csm.Odometer.mean())
```

```python
csm
```

```python
# not reassigned, we can either do csm.Odometer = csm['Odometer].fillna(...) 
# OR, set inplace parameter True, which is False by default
csm.Odometer.fillna(csm.Odometer.mean(),inplace = True)
csm # as we can see, missing values are filled now
```

```python
#dropna function used to drop rows with missing values
csm.dropna(inplace=True)
csm
# if we want to assign it to another df, do not set inplace True as this will change the orignal df
```

```python
csm = pd.read_csv(car_sales_missing_path)
csm_dropped = csm.dropna()
csm
```

```python
csm_dropped
```

```python
seats_column = pd.Series([5,5,5,5,5])
csm['Seats'] = seats_column # adding data example
csm # creation of column from a Series ^
```

```python
# to fill missing columns in seats
csm.Seats.fillna(5,inplace=True)
csm
```

```python
# creating column from a python list
fuel_economy = [7.5,8.5,5.4,9,9.3]
#csm['FuelEcon'] = fuel_economy does not work as length of column does not match length of index error
# we will have to use a list with same length as data frame
fuel_economy.extend([3,4.5,6,3.4,8.9])
csm['FuelEcon'] = fuel_economy
csm
```

```python
csm.Odometer.fillna(csm.Odometer.mean(),inplace=True)
csm['fuelUsed'] = csm.Odometer/100 * csm.FuelEcon
csm
```

```python
csm['no. of wheels'] = 4
csm
```

```python
# axis of row is 0, axis of column is 1
csm.drop('no. of wheels',axis=1,inplace=True)
csm
```

```python
#sample used to shuffle, 
csm_shuffled = csm.sample(frac=1) # frac is the fraction of data to be shuffles
csm_shuffled
```

```python
csm_shuffled.reset_index(inplace=True,drop=True)
csm_shuffled
# used to reset the index column, does not unshuffle the dataframe
```
